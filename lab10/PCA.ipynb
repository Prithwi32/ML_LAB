```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('IRIS.csv')
if 'Id' in df.columns:
    df = df.drop('Id', axis=1)
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
y = df['species'].values
label_encoder = LabelEncoder()
y_encoder = label_encoder.fit_transform(y)
pca = PCA(n_components=2)
X_projected = pca.fit_transform(X)
print("Original Shape: ",X.shape)
print("Transformed Shape:", X_projected.shape)
plt.figure(figsize=(8,6))
plt.scatter(X_projected[:,0], X_projected[:,1], c=y_encoder, cmap='jet')
plt.xlabel("Pricipal Component 1")
plt.ylabel("Pricipal Component 2")
plt.title("PCA of Iris Dataset (from CSV)")
plt.colorbar(label="Encoded Species")
plt.show()
```
